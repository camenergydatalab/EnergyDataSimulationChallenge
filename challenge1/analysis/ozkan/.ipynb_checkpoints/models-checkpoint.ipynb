```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train = pd.read_csv('../../data/training_dataset_500.csv')
test = pd.read_csv('../../data/test_dataset_500.csv')
```

```python
# Looking at the head of data, we have 23 months for each house and will predict month of June
train.head(30)
```

```python
test.head()
```

```python
# House is a critical input here, energy consumption mostly relies on this value
train.groupby('House')['EnergyProduction'].mean()
```

```python
train.groupby('Month')['EnergyProduction'].mean().plot()
plt.show()
```

```python
# Interestingly year is not a categorical variable but a ordinal variable
train.groupby('Year')['EnergyProduction'].mean()
```

```python
train_y = train['EnergyProduction']
test_y = test['EnergyProduction']
train_x = train.drop(columns=['ID', 'Label', 'EnergyProduction'])
test_x = test.drop(columns=['ID', 'Label', 'EnergyProduction'])               
```

```python
# Let's define de MAPE function as it doesn't exit in sklearn library
def MAPE(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
# 1- Trying models with raw data
# 1.1 XGBOOST
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(train_x, train_y)
pred_y = xgb.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('XGBOOST')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# 1.2 LightGBM
from lightgbm import LGBMRegressor
lgb = LGBMRegressor()
lgb.fit(train_x, train_y)
pred_y = lgb.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('LightGBM')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# 1.3 CatBoost
from catboost import CatBoostRegressor
catb = CatBoostRegressor(allow_writing_files=False)
catb.fit(train_x, train_y)
pred_y = catb.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('CatBoost')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# 1.4 Linear Regression
import statsmodels.formula.api as smf
res = smf.glm('EnergyProduction ~ Temperature + Daylight + House + Year + Month', train).fit()
pred_y = res.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('GLM')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# 1.5 SVR
from sklearn import svm
svr = svm.SVR(kernel='linear')
svr.fit(train_x, train_y)
pred_y = svr.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('SVR')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# GLM seems to be quite good as we were able to feed model with categoric House variable. 
# To Feed other models, I will one hot encode "House" variable
# 2- One Hot Encode House var'able
```

```python
# One Hot Encode categorical variables
from sklearn.preprocessing import OneHotEncoder
cat_fields = ['House', 'Year', 'Month']

one_hot_house_train = pd.get_dummies(train_x.House, prefix = 'House')
one_hot_house_test = pd.get_dummies(test_x.House, prefix = 'House')
```

```python
train_x = pd.concat( [train_x, one_hot_house_train], axis = 1)
test_x = pd.concat( [test_x, one_hot_house_test], axis = 1)
```

```python
train_x.drop(columns=['House'], inplace=True)
test_x.drop(columns=['House'], inplace=True)
```

```python
# 2.1 XGBOOST on Categoric Houses
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(train_x, train_y)
pred_y = xgb.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('XGBoost on Categoric Houses')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# 2.2 LightGBM on Categoric Houses
from lightgbm import LGBMRegressor
lgb = LGBMRegressor()
lgb.fit(train_x, train_y)
pred_y = lgb.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('LightGBM on Categoric Houses')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# 2.3 CatBoost on Categoric Houses
from catboost import CatBoostRegressor
catb = CatBoostRegressor(allow_writing_files=False)
catb.fit(train_x, train_y)
pred_y = catb.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('CatBoost on Categoric Houses')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# 2.4 Linear Regression on Categoric Houses
import statsmodels.formula.api as smf
res = smf.glm('EnergyProduction ~ Temperature + Daylight + C(House) + Year + Month', train).fit()
pred_y = res.predict(test)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('GLM on Categoric Houses')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# 2.5 SVR on Categoric Houses
from sklearn import svm
svr = svm.SVR(kernel='linear')
svr.fit(train_x, train_y)
pred_y = svr.predict(test_x)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('SVR on Categoric Houses')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```

```python
# GLM performed well both on raw data and categoric data, I will focus on it from now on
import statsmodels.formula.api as smf
res = smf.glm('EnergyProduction ~ Temperature + Daylight + C(House) + Year + Month', train).fit()
pred_y = res.predict(test)
MAPE(test_y, pred_y)
```

```python
# Let's try to generate a feature where we can increase importance of Month factor
print('Testing different absolute month differences')
for month in train['Month'].unique():
    res = smf.glm('EnergyProduction ~ Temperature + Daylight + C(House) + Year + Month + abs(Month-month)', train).fit()
    pred_y = res.predict(test)
    print('Month: '+str(month), end='\t')
    print('%0.2f' %MAPE(test_y, pred_y))
```

```python
# Building final model
res = smf.glm('EnergyProduction ~ Temperature + Daylight + C(House) + Year + Month + abs(Month-6)', train).fit()
pred_y = res.predict(test)
MAPE(test_y, pred_y)
```

```python
# Visualizing predictions vs real values
import matplotlib.pyplot as plt
plt.scatter(test_y, pred_y, alpha=0.5)
plt.title('Final GLM on Generated Features')
plt.xlabel('Test')
plt.ylabel('Pred')
plt.show()
```
