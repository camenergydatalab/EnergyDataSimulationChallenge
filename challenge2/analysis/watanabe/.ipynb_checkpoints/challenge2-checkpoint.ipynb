```python
# -*-coding:utf-8-*-
%matplotlib inline
import math
import csv
import calendar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import *
from datetime import datetime
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
# step1:Download the data-set total-watt.csv
tw = pd.read_csv("../../data/total_watt.csv",
                 header=None,
                 parse_dates=["Timestamp"], 
                 names=["Timestamp","Energy"], 
                 index_col="Timestamp")

tw.head()
```

```python
# step3:visualise the data-set
f = plt.figure(figsize=(15, 5))
tw.plot(ax=f.gca())
plt.title("The energy consumption per 30mins")
plt.xlabel("Time")
plt.ylabel("The energy consumption(W)")

plt.show()
```

```python
# step4:visualise the data-set as values per day

dr = pd.date_range("2011-04-18","2011-05-24")

day = np.zeros([len(dr)])
night = np.zeros([len(dr)])
total = np.zeros([len(dr)])

for i,one_date in enumerate(dr):
    one_date = one_date.strftime("%Y-%m-%d")
    day[i] = tw[one_date+" 08:00:00":one_date+" 18:00:00"].sum()
    night[i] = tw[one_date+" 18:00:00":one_date].sum()+tw[one_date:one_date+' 8:00:00'].sum()
    total[i] = tw[one_date].sum()

ind = np.arange(len(dr))
plt.figure(figsize=(15, 5))  
plt.bar(ind, total, color="k",alpha=0.2,label="total")
plt.plot(ind,day, color="r",label="daytime", marker="o")
plt.plot(ind,night, color="b",label="night", marker="o")
plt.xticks(ind, [datetime.strftime(i,"%m/%d") for i in dr], rotation=60)
plt.legend(loc="upper left")
plt.title("The energy consumptions per day")
plt.xlabel("Date")
plt.ylabel("The energy consumption(W)")

plt.show()
```

```python
# step6:visualise the clusters
pred = KMeans(n_clusters=3)
pred.fit_predict([[x] for x in total])
labels = pred.labels_

high = labels[total.argmax()]
low = labels[total.argmin()]
color_list = ["#ffa500"]*3
color_list[high] = "#dc143c"
color_list[low] = "#ffff66"
color = [color_list[i] for i in labels]

plt.figure(figsize=(15, 5))    
ind = np.arange(len(dr))

plt.bar(ind, total, color=color,alpha=0.8)
plt.xticks(ind, [datetime.strftime(i,"%m/%d") for i in dr], rotation=60)
plt.title("The energy consumptions per day")
plt.xlabel("Date")
plt.ylabel("The energy consumption(W)")

plt.show()
```

```python
week = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]

a = [0,0,0,0,0,0,0]

plt.figure(figsize=(15, 5))  
for i,e in enumerate(total):
    q = i // 7
    w = i % 7
    a[w] += e
    plt.bar(q*0.3+w*2,e,0.2,color=color[i])

# Separate graph by the day of the week
for x in range(0,12,2):
    plt.axvline(x = x+1.7, linestyle="--", color="k")

plt.xticks(np.arange(0.8,14,2),week)
plt.title("Difference of energy consumption per day of the week")
plt.xlabel("Day of the week")
plt.ylabel("The energy consumption(W)")

plt.show()
```

```python
# generate average energy
b = [0,0,0,0,0,0,0]
b[0] = a[0]/6
b[1] = a[1]/6
b[2] = a[2]/5
b[3] = a[3]/5
b[4] = a[4]/5
b[5] = a[5]/5
b[6] = a[6]/5
for i in range(len(b)):
    b[i] = floor(b[i])
    b[i] = int(b[i])
    b[i] = str(b[i])+"(w)"

# pie chart parameters
plt.style.use('ggplot')
plt.rcParams.update({'font.size':15})
size=(10,5)
col=cm.Spectral(np.arange(len(a))/float(len(a)))
plt.figure(figsize=size,dpi=100)
plt.pie(a,colors=col,counterclock=False,
        startangle=90,labels=b)
plt.subplots_adjust(left=0,right=0.7)
plt.legend(week,fancybox=True,loc='center left',bbox_to_anchor=(0.9,0.5))
plt.axis('equal') 
plt.savefig('figure.png',bbox_inches='tight',pad_inches=0.05)

plt.title("Average energy consumption per day of the week")

plt.show()
```
